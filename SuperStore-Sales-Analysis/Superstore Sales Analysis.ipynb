```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("Downloads/Superstore.csv/Superstore.csv", encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.info()
```

```python
df["Order Year"] = df["Order Date"].dt.year
```

```python
df.info()
```

```python
df["Order Year"].value_counts()
```

```python
sales_by_year = df.groupby(["Order Year"])["Sales"].sum().reset_index()
sales_by_year
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x ="Order Year", y = "Sales"  , data=sales_by_year, marker = "o", sort = False)
```

```python
df = df.drop("Month_Year", axis=1)
```

```python
df.info()
```

```python
df["Month Year"] = df["Order Date"].dt.to_period("M").astype(str)
```

```python
df.info()
```

```python
sales_by_month = df.groupby(["Month Year"])["Sales"].sum().reset_index()
sales_by_month
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="Month Year", y="Sales", data = sales_by_month)
plt.xticks(rotation=45)
plt.show()
```

```python
df["Order Month"] = df["Order Date"].dt.month
```

```python
df.info()
```

```python
agg_month = df.groupby(["Order Year","Order Month"])["Sales"].sum()
```

```python
agg_month = agg_month.reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="Order Month", y="Sales", hue = "Order Year", palette="RdBu" ,data = agg_month)
```

```python
yearly_profits = df.groupby(["Order Year"])["Profit"].sum().reset_index()
yearly_profits
```

```python
sns.lineplot(x="Order Year", y="Profit", data=yearly_profits, marker="o")
```

```python
sales_group = df.groupby(["Order Year", "Category"])["Sales"].sum().reset_index()
sales_group
```

```python
profit_group = df.groupby(["Order Year", "Category"])["Profit"].sum().reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Order Year", y="Sales", hue="Category",palette= "coolwarm", data=sales_group)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Order Year", y="Profit", hue="Category",palette= "coolwarm", data=profit_group)
```
